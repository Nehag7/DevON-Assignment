### Another Way of scraping data from API 

```python
# one way 
payloads = {'exporttype':'bidpriceladder',
            'format':'csv',
            'datefrom':'2020-01-01',
            'dateto':'2020-01-02',
            'submit':1}
response = requests.get(
    'https://www.tennet.org/english/operational_management/export_data.aspx',
    params= payloads)
```

```python
tf = open('devon.csv' , 'wb')
tf.write(response.content)
tf.close()
nh = pd.read_csv('devon.csv')
nh.head()
```
